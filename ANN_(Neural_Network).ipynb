```python
#standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the data
df=pd.read_csv('HEMQ.csv')
```

```python
#Neural network need data completion. So follow imputation or deletion
NA_names= ['MORTDUE', 'VALUE', 'YOJ', 'DEROG',
'DELINQ', 'CLAGE', 'NINQ', 'CLNO', 'DEBTINC']
df[NA_names]=df[NA_names].fillna(df[NA_names].mean())
```

```python
#categorial variables imputation
df['REASON']=df['REASON'].fillna(df['REASON'].mode()[0])
df['JOB']=df['JOB'].fillna(df['JOB'].mode()[0])
```

```python
df.describe(include='all')
```

```python
X=pd.concat([df.iloc[:,1:4],df.iloc[:,6:13]],axis=1)
names=X.columns
```

```python
#standardization of variables for neural networks
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X[names]=scaler.transform(X)
X.describe(include='all')
```

```python
dummy_data=pd.get_dummies(df[['REASON','JOB']],drop_first=True)
X=pd.concat([X,dummy_data],axis=1)
X.head()
```

```python
y=df['BAD']
```

```python
#data partition
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
#build neural network
from sklearn.neural_network import MLPClassifier
ANN = MLPClassifier(hidden_layer_sizes=(5,3),
                    max_iter=300,
                    random_state=0)
ANN.fit(X_train,y_train)
```

```python
#coefficients for the first hidden layer
print(ANN.coefs_[0])
```

```python
#coefficients for the 2nd hidden layer. Output of 1st hidden layer = input to 2nd layer
print(ANN.coefs_[1])
```

```python
#coefficients of the output layer. Output of 2nd hidden layer = input of last year
print(ANN.coefs_[2])
```

```python
#compare the model to the validation set
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
prediction=ANN.predict(X_val)
confusion_matrix(y_val,prediction)
```

```python
ANN.score(X_val,y_val)
```

```python
prediction_prob=ANN.predict_proba(X_val)
roc_auc_score(y_val,prediction_prob[:,1])
```

```python
from sklearn.inspection import permutation_importance

perm = permutation_importance(ANN,
                              X_val,
                              y_val,
                              n_repeats=30,
                              random_state=0,
                              scoring='roc_auc')
```

```python
perm_series = pd.Series(data=perm.importances_mean,
                        index=X_val.columns.values).sort_values(ascending=False)
display(perm_series.head(15))
```
